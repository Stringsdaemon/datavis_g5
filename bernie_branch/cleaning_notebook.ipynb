```python
import pandas as pd
```

```python
data = pd.read_csv('google_clean.csv')
data
```

```python
data['Last_Updated'] = pd.to_datetime(
    data[['year', 'month', 'Day']],  # Select the 3 columns
    format='%Y-%m-%d',             # Define the input format
    errors='coerce'                # Handle invalid dates as NaT (Not-a-Time)
)
```

```python
data = data.drop('Unnamed: 0', axis=1)
```

```python
data
```

```python
data.to_csv('google_clean_v2.csv', index=False)
```
